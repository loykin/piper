```python
import json
import os

from support.math_utils import scaled_values
from tinyrecord import record

features = record(scaled_values(multiplier))
features["multiplier"] = multiplier
output_path = os.path.join(os.environ["PIPER_OUTPUT_DIR"], "features.json")
with open(output_path, "w", encoding="utf-8") as dst:
    json.dump(features, dst)
print(json.dumps(features, sort_keys=True))
```
