```python
multiplier = 2
```

```python
import json
import os

from support.math_utils import scaled_values
from tinyrecord import record

seed_path = os.path.join(os.environ["PIPER_INPUT_DIR"], "seed", "seed.json")
with open(seed_path, encoding="utf-8") as src:
    seed = json.load(src)
if seed["multiplier"] != multiplier:
    raise RuntimeError(f"seed multiplier mismatch: {seed!r}")

features = record(scaled_values(multiplier))
if features["values"] != [value * multiplier for value in seed["values"]]:
    raise RuntimeError(f"unexpected seed values: {seed!r}")
features["multiplier"] = multiplier
output_path = os.path.join(os.environ["PIPER_OUTPUT_DIR"], "features.json")
with open(output_path, "w", encoding="utf-8") as dst:
    json.dump(features, dst)
print(json.dumps(features, sort_keys=True))
```
